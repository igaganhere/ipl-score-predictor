```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


# OneHot Encode categorical columns
columnTransformer = ColumnTransformer([
    ('encoder', OneHotEncoder(), ['bat_team', 'bowl_team'])
], remainder='passthrough')

# Fit and transform
df = columnTransformer.fit_transform(df)

# Restore column names
col_names = columnTransformer.get_feature_names_out()
df = pd.DataFrame(df.toarray(), columns=col_names)

print(df.shape)
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace('encoder__', '')
df.columns = df.columns.str.replace('remainder__', '')
```

```python
df
```

```python
x = df.drop(['total'], axis=1)
y = df['total']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, shuffle=True,random_state=42)
```

```python
models=dict()
```

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor()

tree.fit(x_train, y_train)
```

```python
train_score_tree = str(tree.score(x_train, y_train) * 100)
test_score_tree = str(tree.score(x_test, y_test) * 100)
print(f'Train Score : {train_score_tree[:5]}%\nTest Score : {test_score_tree[:5]}%')
models["tree"] = test_score_tree
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import root_mean_squared_error as rmse
```

```python
print(f'MAE={mae(y_test,tree.predict(x_test))}')
print(f'MSE={mse(y_test,tree.predict(x_test))}')
print(f'RMSE={rmse(y_test,tree.predict(x_test))}')
```

```python
def score_predict(batting_team, bowling_team, runs, wickets, overs, runs_last_5, wickets_last_5, model=forest):
  prediction_array = []
  batting_teams = [
    'Chennai Super Kings', 'Deccan Chargers', 'Delhi Daredevils',
    'Gujarat Lions', 'Kings XI Punjab', 'Kochi Tuskers Kerala',
    'Kolkata Knight Riders', 'Mumbai Indians', 'Pune Warriors',
    'Rajasthan Royals', 'Rising Pune Supergiants',
    'Royal Challengers Bangalore', 'Sunrisers Hyderabad'
  ]

  bowling_teams = [
    'Chennai Super Kings', 'Deccan Chargers', 'Delhi Daredevils',
    'Gujarat Lions', 'Kings XI Punjab', 'Kochi Tuskers Kerala',
    'Kolkata Knight Riders', 'Mumbai Indians', 'Pune Warriors',
    'Rajasthan Royals', 'Rising Pune Supergiant', 'Rising Pune Supergiants',
    'Royal Challengers Bangalore', 'Sunrisers Hyderabad'
]
# Batting Team Encoding
  batting_array = [1 if team == batting_team else 0 for team in batting_teams]

# Bowling Team Encoding
  bowling_array = [1 if team == bowling_team else 0 for team in bowling_teams]
  prediction_array = []
  prediction_array = batting_array + bowling_array
  prediction_array = prediction_array + [runs, wickets, overs, runs_last_5, wickets_last_5]
  prediction_array = np.array([prediction_array])
  pred = model.predict(prediction_array)
  return int(round(pred[0]))
```

```python
batting_team="Chennai Super Kings"
bowling_team="Mumbai Indians"
score = score_predict(batting_team, bowling_team, overs=18.0, runs=150, wickets=4, runs_last_5=57, wickets_last_5=1)
print(f'Predicted Score : {score} || Actual Score : 172')
```
